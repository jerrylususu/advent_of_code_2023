```python
with open("day14.txt", "r") as f:
    data = f.read()
```

```python
data
```

```python
pattern = []
for line in data.split("\n"):
    if line.strip() == "":
        continue
    row = []
    for char in line:
        row.append(char)
    pattern.append(row)
```

```python
pattern
```

```python
def find_most_up_non_empty(pattern, y, x):
    current_y = y - 1
    while current_y >= 0:
        # print("current_y", current_y, pattern[current_y][x])
        if pattern[current_y][x] == '.':
            current_y -= 1
        else:
            break
    return current_y + 1 # leave some space

    # should not be here
    raise Exception("should not be here")
```

```python
find_most_up_non_empty(pattern, 1, 0)
```

```python
for row_idx, row in enumerate(pattern):
    for col_idx, char in enumerate(row):
        if char == 'O':
            next_y_pos = find_most_up_non_empty(pattern, row_idx, col_idx)
            if next_y_pos != row_idx:
                pattern[next_y_pos][col_idx] = 'O'
                pattern[row_idx][col_idx] = '.'
```

```python
pattern
```

```python
total = 0
for row_idx, row in enumerate(pattern):
    row_weight = len(pattern) - row_idx
    current_row_count = sum([1 for char in row if char == 'O'])
    total += row_weight * current_row_count
total
```
