```python
def calculate_sum_from_file(file_path):
    total_sum = 0
    with open(file_path, 'r') as file:
        for line in file:
            first_digit = int(next(filter(str.isdigit, line)))
            last_digit = int(next(filter(str.isdigit, line[::-1])))
            combined_number = int(str(first_digit) + str(last_digit))
            total_sum += combined_number
    return total_sum

file_path = 'day1a.txt'
result = calculate_sum_from_file(file_path)
print(result)
```
