```python
from functools import cache

@cache
def count_solutions(remaining_pattern, remaining_value, consecutive_count=0) -> int:
    # print("remaining_pattern", remaining_pattern, "remaining_value", remaining_value, "consecutive_count", consecutive_count, "full", full)
    if remaining_pattern == "" and len(remaining_value) == 0:
        # ok
        if consecutive_count == 0:
            # print("ok1", full)
            return 1
        else:
            return 0
    if remaining_pattern == "" and len(remaining_value) > 0:
        remaining_values = [int(i) for i in remaining_value.split(",")]
        if remaining_values[0] == consecutive_count and len(remaining_values) == 1:
            # print("ok2", full)
            return 1

        # print("hit 3")
        # missing some value, invalid
        return 0

    if len(remaining_value) == 0:
        remaining_values = []
    else:
        remaining_values = [int(i) for i in remaining_value.split(",")]

    char, further = remaining_pattern[0], remaining_pattern[1:]

    if char == ".":
        # backtrack for #
        if consecutive_count > 0:
            if len(remaining_values) == 0:
                # print("hit 1")
                return 0
            if remaining_values[0] != consecutive_count:
                # print("hit 2")
                return 0
            remaining_values = remaining_values[1:]
            remaining_value = ",".join([str(i) for i in remaining_values])
        return count_solutions(further, remaining_value, 0)
    elif char == "#":
        return count_solutions(further, remaining_value, consecutive_count + 1)
        # if len(remaining_values) == 0:
        #     # invalid
        #     return 0
        # remaining_values[0] -= 1
        # if remaining_values[0] == 0:
        #     remaining_values = remaining_values[1:]
        # return count_solutions(further, ",".join([str(i) for i in remaining_values]), full)

    elif char == "?":
        # print("inner further", further)
        return count_solutions("." + further, remaining_value, consecutive_count) + \
            count_solutions("#" + further, remaining_value, consecutive_count)
```

```python
count_solutions(patterns[0], values[0])
```

```python
results = []
for p,v in zip(patterns, values):
    solutions = count_solutions(p, v)
    print(p, v, solutions)
    results.append(solutions)
results
```
