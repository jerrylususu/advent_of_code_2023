```python
with open("day22.txt", "r") as f:
    lines = [i.strip() for i in f.readlines()]
```

```python
def parse_line(line):
    start_coord_str, finish_coord_str = line.split("~")
    start_x, start_y, start_z = start_coord_str.split(",")
    finish_x, finish_y, finish_z = finish_coord_str.split(",")
    return ((int(start_x), int(start_y), int(start_z)), (int(finish_x), int(finish_y), int(finish_z)))
```

```python
bricks = [parse_line(i) for i in lines]
```

```python
bricks
```

```python
def get_all_points_in_projection(brick):
    start, finish = brick
    start_x, start_y, start_z = start
    finish_x, finish_y, finish_z = finish
    return [(x, y) for x in range(start_x, finish_x + 1) for y in range(start_y, finish_y + 1)]
```

```python
min_x = min([i[0][0] for i in bricks])
min_y = min([i[0][1] for i in bricks])
max_x = max([i[1][0] for i in bricks])
max_y = max([i[1][1] for i in bricks])
min_x, min_y, max_x, max_y
```

```python
# simulate the fall...
# ground at z=0, first layer at z=1, all inclusive

current_ground = []
for x in range(min_x, max_x + 1):
    arr = []
    for y in range(min_y, max_y + 1):
        arr.append(0)
    current_ground.append(arr)

landed_bricks = []

# 1. sort bricks by start_z
bricks.sort(key=lambda x: x[0][2])

# 2. for each brick, find the projection area, and the largest z on that area
for brick in bricks:
    projection_points = get_all_points_in_projection(brick)
    max_z_on_projection = 0
    for point in projection_points:
        max_z_on_projection = max(max_z_on_projection, current_ground[point[0]][point[1]])
    # now we landed
    # lower_z
    new_z = max_z_on_projection + 1

    # and we shall update the finished brick
    landed_bricks.append((brick, new_z))
    current_brick_height = brick[1][2] - brick[0][2] + 1

    # and we need to update the ground
    for point in projection_points:
        current_ground[point[0]][point[1]] = new_z + current_brick_height - 1    

```

```python
base_landed_bricks = landed_bricks
```

```python
# now try to remove?

ok_remove = []
for idx, target_brick in enumerate(bricks):
    current_ground = []
    for x in range(min_x, max_x + 1):
        arr = []
        for y in range(min_y, max_y + 1):
            arr.append(0)
        current_ground.append(arr)

    landed_bricks = []

    # 1. sort bricks by start_z
    bricks.sort(key=lambda x: x[0][2])

    # 2. for each brick, find the projection area, and the largest z on that area
    for brick in bricks:
        if brick == target_brick:
            continue
        projection_points = get_all_points_in_projection(brick)
        max_z_on_projection = 0
        for point in projection_points:
            max_z_on_projection = max(max_z_on_projection, current_ground[point[0]][point[1]])
        # now we landed
        # lower_z
        new_z = max_z_on_projection + 1

        # and we shall update the finished brick
        landed_bricks.append((brick, new_z))
        current_brick_height = brick[1][2] - brick[0][2] + 1

        # and we need to update the ground
        for point in projection_points:
            current_ground[point[0]][point[1]] = new_z + current_brick_height - 1    
    
    print(idx)

    # 3. check if that's ok
    moved = 0
    for brick, current_z in landed_bricks:
        # first find the brick and base_z in base_landed_bricks
        for base_brick, base_z in base_landed_bricks:
            if base_brick == brick:
                if current_z != base_z:
                    moved += 1
    if moved == 0:
        ok_remove.append(idx)
```
