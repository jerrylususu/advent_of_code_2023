```python
lines = data.split()
```

```python
def number_of_items_less_than_n(item_list, n):
    count = 0
    for item in item_list:
        if item < n:
            count += 1
    return count
```

```python
# 1. find all the galaxies
galaxies = []
for row_idx, row in enumerate(lines):
    for col_idx, val in enumerate(row):
        if val == '#':
            galaxies.append((row_idx, col_idx))

# 2. find row/columns that does not has any galaxy
no_row = []
no_col = []
for row_idx, row in enumerate(lines):
    if "#" not in row:
        no_row.append(row_idx)

for col_idx in range(len(lines[0])):
    if "#" not in [row[col_idx] for row in lines]:
        no_col.append(col_idx)

# 3. do expand
expanded_galaxies = []
for row, col in galaxies:
    more_row = number_of_items_less_than_n(no_row, row)
    more_col = number_of_items_less_than_n(no_col, col)
    expanded_galaxies.append((row + more_row, col + more_col))

# 4. calculate pair wise manhattan distance
total_distance = 0
# iterate get all pairs
for i in range(len(expanded_galaxies)):
    for j in range(i+1, len(expanded_galaxies)):
        total_distance += abs(expanded_galaxies[i][0] - expanded_galaxies[j][0]) + abs(expanded_galaxies[i][1] - expanded_galaxies[j][1])

total_distance
```
