```python
with open("day17.txt", "r") as f:
    lines = f.readlines()

pattern = []
for line in lines:
    pattern.append([int(char) for char in line.strip()])
# make pattern a tuple
pattern = tuple(tuple(row) for row in pattern)
```

```python
import heapq

def shortest_path(pattern):
    # Define the possible directions
    directions = [(0, 1), (0, -1), (1, 0), (-1, 0), (0, 4), (0, -4), (4, 0), (-4, 0)]
    
    # Initialize the priority queue
    # element: weight, current position, last direction, straight steps, path
    queue = [(0, (0, 0), None, 0, [(0, 0)])]


    
    # Set to keep track of visited nodes
    visited = set()
    
    while queue:
        # Pop the path with smallest weight
        weight, (x, y), last_direction, straight_steps, path = heapq.heappop(queue)
        
        # If this is the destination, return the weight and the path
        if (x, y) == (len(pattern) - 1, len(pattern[0]) - 1):
            return weight, path
        
        # If we have visited this node before with the same direction and steps, skip it
        if ((x, y), last_direction, straight_steps) in visited:
            continue
        
        visited.add(((x, y), last_direction, straight_steps))
        
        # Try all possible directions
        for dx, dy in directions:
            real_step = abs(dx) + abs(dy)
            new_x, new_y = x + dx, y + dy

            # check if the new position is already in the path
            if (new_x, new_y) in path:
                continue
            
            # Check if the new position is valid
            if 0 <= new_x < len(pattern) and 0 <= new_y < len(pattern[0]):
                new_direction = (dx // real_step, dy // real_step)
                
                # Check if we are going in the same direction
                if new_direction == last_direction:
                    new_steps = straight_steps + real_step
                else:
                    new_steps = real_step

                # fix path
                middle_points = []
                middle_weights = []
                for i in range(1, real_step + 1):
                    middle_points.append((x + i * dx // real_step, y + i * dy // real_step))
                    middle_weights.append(pattern[x + i * dx // real_step][y + i * dy // real_step])
                # if x == 0 and y == 4:
                #     print(new_direction, (new_x, new_y), real_step, middle_weights)

                # If we have not exceeded the limit of straight steps, add the new path to the queue
                if 4 <= new_steps and new_steps <= 10:
                    # if x == 0 and y == 4:
                    #     print("accepted", (x, y) ,(new_x, new_y), "nd", new_direction, "ns", new_steps, path)
                    new_weight = weight + sum(middle_weights)
                    new_path = path + middle_points
                    heapq.heappush(queue, (new_weight, (new_x, new_y), new_direction, new_steps, new_path))
                # else:
                    # if x == 0 and y == 4:
                    #     print("rejected", (x, y) ,(new_x, new_y), "nd", new_direction, "ns", new_steps, path)
    # If no valid path is found, return -1 and an empty path
    return -1, []
```

```python
total_weight, path = shortest_path(pattern)
total_weight, path
```
