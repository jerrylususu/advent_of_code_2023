```python
with open("day18.txt", "r") as f:
    lines = f.readlines()
    lines = [i.strip() for i in lines]
```

```python
def parse_line(line):
    line = line.split(" ")
    direction, distance = line[0], int(line[1])
    color = line[2].replace("(", "").replace(")", "")
    return direction, distance, color
```

```python
lines = [parse_line(i) for i in lines]
```

```python
cur_row, cur_col = 0, 0
passed_coords = []

for line in lines:
    direction, distance, color = line
    if direction == "D":
        for i in range(distance):
            cur_row += 1
            passed_coords.append((cur_row, cur_col))
    elif direction == "U":
        for i in range(distance):
            cur_row -= 1
            passed_coords.append((cur_row, cur_col))
    elif direction == "L":
        for i in range(distance):
            cur_col -= 1
            passed_coords.append((cur_row, cur_col))
    elif direction == "R":
        for i in range(distance):
            cur_col += 1
            passed_coords.append((cur_row, cur_col))
```

```python
passed_coords
```

```python
border_up_limit = max([i[0] for i in passed_coords])
border_down_limit = min([i[0] for i in passed_coords])
border_left_limit = min([i[1] for i in passed_coords])
border_right_limit = max([i[1] for i in passed_coords])
```

```python
border_rows = border_up_limit - border_down_limit + 1
border_cols = border_right_limit - border_left_limit + 1
```

```python
grid = [["." for i in range(border_cols)] for j in range(border_rows)]
```

```python
for coord in passed_coords:
    # with offsets
    new_row = coord[0] - border_down_limit
    new_col = coord[1] - border_left_limit
    grid[new_row][new_col] = "#"
```

```python
def get_inside_points_using_flood_fill(grid):
    inside_points = []
    rows = len(grid)
    cols = len(grid[0])
    
    to_be_filled_queue = []
    already_filled = set()
    
    to_be_filled_queue.append((383,85)) # a known inner point

    while len(to_be_filled_queue) > 0:
        row, col = to_be_filled_queue.pop(0)
        if (row, col) in already_filled:
            continue
        already_filled.add((row, col))
        
        if grid[row][col] == ".":
            inside_points.append((row, col))
            if row + 1 < rows and (row + 1, col) not in already_filled:
                to_be_filled_queue.append((row + 1, col))
            if row - 1 >= 0 and (row - 1, col) not in already_filled:
                to_be_filled_queue.append((row - 1, col))
            if col + 1 < cols and (row, col + 1) not in already_filled:
                to_be_filled_queue.append((row, col + 1))
            if col - 1 >= 0 and (row, col - 1) not in already_filled:
                to_be_filled_queue.append((row, col - 1))
    
    return inside_points
```

```python
# check if there is any intersection between flood fill results and passed coord
flood_fill_results = get_inside_points_using_flood_fill(grid)

# get the union
union = set(flood_fill_results).union(set(passed_coords))
len(union)
```

```python
def visualize_inside_points(grid, inside_points):
    grid = json.loads(json.dumps(grid))
    for point in inside_points:
        grid[point[0]][point[1]] = "O"
    for row in grid:
        print("".join(row))
```

```python
visualize_inside_points(grid, get_inside_points_using_flood_fill(grid))
```
