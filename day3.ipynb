```python
# 1. in each row of data, find out all numbers (consecutive digits)
number_with_position = []
for row_idx, row in enumerate(data.split("\n")):
    number = ""
    col_idx = 0
    for col_idx, char in enumerate(row):
        if char.isdigit():
            number += char
        else:
            if number:
                starting_col_idx = col_idx - len(number) 
                number_with_position.append((number, row_idx, starting_col_idx, col_idx - 1))
                number = ""
    if number:
        starting_col_idx = len(row) - len(number) 
        number_with_position.append((number, row_idx, starting_col_idx, len(row) - 1))

# A gear is any * symbol that is adjacent to exactly two part numbers.
gears_to_numbers = {}
for number, row_idx, starting_col_idx, ending_col_idx in number_with_position:
    surrounding_box = []
    gear_locations = []
    for i in range(row_idx - 1, row_idx + 2):
        for j in range(starting_col_idx - 1, ending_col_idx + 2):
            if i < 0 or j < 0:
                continue
            try:
                surrounding_box.append(data.split("\n")[i][j])
                if (data.split("\n")[i][j] == '*'):
                    # gear!
                    gear_locations.append((i, j))
            except IndexError:
                continue
    if any(not char.isdigit() and char != "." for char in surrounding_box) and any(char == '*' for char in surrounding_box):
        for gear_loc in gear_locations:
            if gear_loc not in gears_to_numbers:
                gears_to_numbers[gear_loc] = []
            gears_to_numbers[gear_loc].append((number, row_idx, starting_col_idx, ending_col_idx))


# # 3. if yes, then it's a found number and record that into a list
# final = sum(int(number) for number, _, _, _ in found_numbers)
```

```python
gear_ratios = []
for gear_loc, numbers in gears_to_numbers.items():
    if len(numbers) == 2:
        gear_ratios.append(int(numbers[0][0])* int(numbers[1][0]))
sum(gear_ratios)
```
