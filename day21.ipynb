```python
with open("day21.txt", "r") as f:
    lines = [line.strip() for line in f.readlines()]
    grid = []
    for line in lines:
        grid.append([c for c in line])
```

```python
grid
```

```python
def locate_start(grid):
    for row in range(len(grid)):
        for col in range(len(grid[row])):
            if grid[row][col] == "S":
                return (row, col)
    raise Exception("No start found")
```

```python
start = locate_start(grid)
start
```

```python
def get_next_locations(grid, cur_row, cur_col):
    locations = []
    if cur_row > 0 and grid[cur_row - 1][cur_col] != "#":
        locations.append((cur_row - 1, cur_col))
    if cur_row < len(grid) - 1 and grid[cur_row + 1][cur_col] != "#":
        locations.append((cur_row + 1, cur_col))
    if cur_col > 0 and grid[cur_row][cur_col - 1] != "#":
        locations.append((cur_row, cur_col - 1))
    if cur_col < len(grid[cur_row]) - 1 and grid[cur_row][cur_col + 1] != "#":
        locations.append((cur_row, cur_col + 1))
    return locations
```

```python
current_locations = []
next_locations = []
last_next_locations = []

current_locations.append(start)

total_steps = 64
for i in range(total_steps):
    while len(current_locations) > 0:
        cur_row, cur_col = current_locations.pop()

        nexts = get_next_locations(grid, cur_row, cur_col)
        for next in nexts:
            if next not in next_locations:
                next_locations.append(next)
    # print("run: ", i, next_locations)

    last_next_locations = next_locations
    # update
    current_locations.extend(next_locations)
    next_locations = []

len(last_next_locations)
```
