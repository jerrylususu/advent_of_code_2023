```python
def find_most_up_non_empty(pattern, y, x):
    current_y = y - 1
    while current_y >= 0:
        # print("current_y", current_y, pattern[current_y][x])
        if pattern[current_y][x] == '.':
            current_y -= 1
        else:
            break
    return current_y + 1 # leave some space

    # should not be here
    raise Exception("should not be here")
```

```python
def find_most_down_non_empty(pattern, y, x):
    current_y = y + 1
    while current_y < len(pattern):
        # print("current_y", current_y, pattern[current_y][x])
        if pattern[current_y][x] == '.':
            current_y += 1
        else:
            break
    return current_y - 1 # leave some space

    # should not be here
    raise Exception("should not be here")
```

```python
def find_most_left_non_empty(pattern, y, x):
    current_x = x - 1
    while current_x >= 0:
        # print("current_x", current_x, pattern[y][current_x])
        if pattern[y][current_x] == '.':
            current_x -= 1
        else:
            break
    return current_x + 1 # leave some space

    # should not be here
    raise Exception("should not be here")
```

```python
def find_most_right_non_empty(pattern, y, x):
    current_x = x + 1
    while current_x < len(pattern[y]):
        # print("current_x", current_x, pattern[y][current_x])
        if pattern[y][current_x] == '.':
            current_x += 1
        else:
            break
    return current_x - 1 # leave some space

    # should not be here
    raise Exception("should not be here")
```

```python
def serialize_pattern(pattern):
    result = ""
    for row in pattern:
        for char in row:
            result += char
    return result
```

```python
def count_pattern(pattern):
    total = 0
    for row_idx, row in enumerate(pattern):
        row_weight = len(pattern) - row_idx
        current_row_count = sum([1 for char in row if char == 'O'])
        total += row_weight * current_row_count
    return total
```

```python
pattern = []
for line in data.split("\n"):
    if line.strip() == "":
        continue
    row = []
    for char in line:
        row.append(char)
    pattern.append(row)
```

```python
seq = [find_most_up_non_empty, find_most_left_non_empty, find_most_down_non_empty, find_most_right_non_empty]
seq_type = ["y", "x", "y", "x"]
iter_order = [0, 1, 2, 3]

previous_patterns = []

for idx in range(100000):
    current_method = seq[idx % len(seq)]
    current_type = seq_type[idx % len(seq_type)]
    current_iter_order = iter_order[idx % len(iter_order)]
    
    if current_iter_order == 0 or current_iter_order == 1:
        for row_idx, row in enumerate(pattern):
            for col_idx, char in enumerate(row):
                if char == 'O':
                    next_pos = current_method(pattern, row_idx, col_idx)
                    if current_type == "y":
                        if next_pos != row_idx:
                            pattern[next_pos][col_idx] = 'O'
                            pattern[row_idx][col_idx] = '.'
                    elif current_type == "x":
                        if next_pos != col_idx:
                            pattern[row_idx][next_pos] = 'O'
                            pattern[row_idx][col_idx] = '.'
    elif current_iter_order == 2 or current_iter_order == 3:
        for row_idx in range(len(pattern)-1, -1, -1):
            for col_idx in range(len(pattern[row_idx])-1, -1, -1):
                char = pattern[row_idx][col_idx]
                if char == 'O':
                    next_pos = current_method(pattern, row_idx, col_idx)
                    if current_type == "y":
                        if next_pos != row_idx:
                            pattern[next_pos][col_idx] = 'O'
                            pattern[row_idx][col_idx] = '.'
                    elif current_type == "x":
                        if next_pos != col_idx:
                            pattern[row_idx][next_pos] = 'O'
                            pattern[row_idx][col_idx] = '.'
    
    if idx % 4 == 3:
        serialized_pattern = serialize_pattern(pattern)
        count = count_pattern(pattern)
        serialized_tuple = (serialized_pattern, current_iter_order, count)
        # ??? the repeat is not simple?
        print(idx, count)
        if previous_patterns.count(serialized_tuple) == 3:
            cycle_indies = [i for i, x in enumerate(previous_patterns) if x == serialized_tuple]
            cycle_length = cycle_indies[1] - cycle_indies[0]
            cycle_start = cycle_indies[0]
            print("found at ", idx, serialized_tuple, "cycle", cycle_indies)
            break
        else:
            previous_patterns.append(serialized_tuple)
    # print(pattern)
print(idx, count)
```
