```python
# In Camel Cards, you get a list of hands, and your goal is to order them based on the strength of each hand. A hand consists of five cards labeled one of A, K, Q, J, T, 9, 8, 7, 6, 5, 4, 3, or 2. The relative strength of each card follows this order, where A is the highest and 2 is the lowest.

# Every hand is exactly one type. From strongest to weakest, they are:
# Five of a kind, where all five cards have the same label: AAAAA
# Four of a kind, where four cards have the same label and one card has a different label: AA8AA
# Full house, where three cards have the same label, and the remaining two cards share a different label: 23332
# Three of a kind, where three cards have the same label, and the remaining two cards are each different from any other card in the hand: TTT98
# Two pair, where two cards share one label, two other cards share a second label, and the remaining card has a third label: 23432
# One pair, where two cards share one label, and the other three cards have a different label from the pair and each other: A23A4
# High card, where all cards' labels are distinct: 23456

from collections import Counter

hand_type_order = ["Five of a kind", "Four of a kind", "Full house", "Three of a kind", "Two pair", "One pair", "High card"]

def get_hand_type(hand: str):
    cnt = Counter([i for i in hand])
    cnt_dict = dict(cnt)
    # sort by values
    cnt_dict = dict(sorted(cnt_dict.items(), key=lambda item: item[1], reverse=True))
    # cnt_dict as a list of tuples
    card_counts = list(cnt_dict.items())

    if len(card_counts) == 1 and card_counts[0][1] == 5:
        return "Five of a kind"
    elif len(card_counts) == 2 and card_counts[0][1] == 4 and card_counts[1][1] == 1:
        return "Four of a kind"
    elif len(card_counts) == 2 and card_counts[0][1] == 3 and card_counts[1][1] == 2:
        return "Full house"
    elif len(card_counts) == 3 and card_counts[0][1] == 3 and card_counts[1][1] == 1 and card_counts[2][1] == 1:
        return "Three of a kind"
    elif len(card_counts) == 3 and card_counts[0][1] == 2 and card_counts[1][1] == 2 and card_counts[2][1] == 1:
        return "Two pair"
    elif len(card_counts) == 4 and card_counts[0][1] == 2 and card_counts[1][1] == 1 and card_counts[2][1] == 1 and card_counts[3][1] == 1:
        return "One pair"
    else:
        return "High card"
```

```python
from functools import cmp_to_key

def compare_hands(hand1: str, hand2: str):
    # first check by type
    hand1_type = get_hand_type(hand1)
    hand2_type = get_hand_type(hand2)
    hand1_type_idx = hand_type_order.index(hand1_type)
    hand2_type_idx = hand_type_order.index(hand2_type)
    if hand1_type_idx > hand2_type_idx:
        return 1
    elif hand1_type_idx < hand2_type_idx:
        return -1
    else:
        # same type, check cards
        hand1_cards = [i for i in hand1]
        hand2_cards = [i for i in hand2]
        # # sort by card order
        # hand1_cards = sorted(hand1_cards, key=cmp_to_key(lambda x, y: card_order.index(x) - card_order.index(y)))
        # hand2_cards = sorted(hand2_cards, key=cmp_to_key(lambda x, y: card_order.index(x) - card_order.index(y)))
        # compare cards one by one
        for i in range(len(hand1_cards)):
            if card_order.index(hand1_cards[i]) > card_order.index(hand2_cards[i]):
                return 1
            elif card_order.index(hand1_cards[i]) < card_order.index(hand2_cards[i]):
                return -1
            else:
                continue
        return 0
```

```python
from collections import Counter

hand_type_order = ["Five of a kind", "Four of a kind", "Full house", "Three of a kind", "Two pair", "One pair", "High card"]

def get_hand_type(hand: str):
    cnt = Counter([i for i in hand])
    cnt_dict = dict(cnt)
    # sort by values
    cnt_dict = dict(sorted(cnt_dict.items(), key=lambda item: item[1], reverse=True))
    # cnt_dict as a list of tuples
    card_counts = list(cnt_dict.items())

    if len(card_counts) == 1 and card_counts[0][1] == 5:
        return "Five of a kind"
    elif len(card_counts) == 2 and card_counts[0][1] == 4 and card_counts[1][1] == 1:
        return "Four of a kind"
    elif len(card_counts) == 2 and card_counts[0][1] == 3 and card_counts[1][1] == 2:
        return "Full house"
    elif len(card_counts) == 3 and card_counts[0][1] == 3 and card_counts[1][1] == 1 and card_counts[2][1] == 1:
        return "Three of a kind"
    elif len(card_counts) == 3 and card_counts[0][1] == 2 and card_counts[1][1] == 2 and card_counts[2][1] == 1:
        return "Two pair"
    elif len(card_counts) == 4 and card_counts[0][1] == 2 and card_counts[1][1] == 1 and card_counts[2][1] == 1 and card_counts[3][1] == 1:
        return "One pair"
    else:
        return "High card"

def get_hand_type_with_J(hand: str):
    if "J" not in hand:
        return get_hand_type(hand)
    # special case:
    # J is a wild card, and can be used to represent any other card in the deck.
    # this function should return the highest rank
    
    index_of_J_s = []
    for idx, card in enumerate(hand):
        if card == "J":
            index_of_J_s.append(idx)
    
    # replace J with all possible cards
    hands_with_j = [hand]
    hands_without_j = []
    while len(hands_with_j) > 0 :
        original_hand = hands_with_j.pop()
        to_modify_hand = [i for i in original_hand]
        # print("to mod", to_modify_hand)
        for card in card_order:
            if card == 'J':
                continue
            for idx in index_of_J_s:
                if original_hand[idx] == "J":
                    to_modify_hand[idx] = card
                    if "J" not in to_modify_hand:
                        hands_without_j.append("".join(to_modify_hand))
                    else:
                        hands_with_j.append("".join(to_modify_hand))
                else:
                    continue
        break
    
    # find the highest hand
    # print(hands_without_j)
    # print(hands_with_j)
    sorted_results = sorted(hands_without_j, key=cmp_to_key(lambda x, y: compare_hands(x, y)))
    return get_hand_type(sorted_results[0])
```

```python
from functools import cmp_to_key

def compare_hands_with_J(hand1: str, hand2: str):
    # first check by type
    hand1_type = get_hand_type_with_J(hand1)
    hand2_type = get_hand_type_with_J(hand2)
    hand1_type_idx = hand_type_order.index(hand1_type)
    hand2_type_idx = hand_type_order.index(hand2_type)
    if hand1_type_idx > hand2_type_idx:
        return 1
    elif hand1_type_idx < hand2_type_idx:
        return -1
    else:
        # same type, check cards
        hand1_cards = [i for i in hand1]
        hand2_cards = [i for i in hand2]
        # # sort by card order
        # hand1_cards = sorted(hand1_cards, key=cmp_to_key(lambda x, y: card_order.index(x) - card_order.index(y)))
        # hand2_cards = sorted(hand2_cards, key=cmp_to_key(lambda x, y: card_order.index(x) - card_order.index(y)))
        # compare cards one by one
        for i in range(len(hand1_cards)):
            if card_order.index(hand1_cards[i]) > card_order.index(hand2_cards[i]):
                return 1
            elif card_order.index(hand1_cards[i]) < card_order.index(hand2_cards[i]):
                return -1
            else:
                continue
        return 0
```
