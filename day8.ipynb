```python
import math

def find_lcm(numbers):
    lcm = numbers[0]
    for i in range(1, len(numbers)):
        lcm = lcm * numbers[i] // math.gcd(lcm, numbers[i])
    return lcm

numbers = [33, 72, 19, 24, 12, 53]
lcm = find_lcm(numbers)
print("The least common multiple is:", lcm)
```
