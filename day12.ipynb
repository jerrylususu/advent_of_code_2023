```python
import itertools
```

```python
def build_value_from_pattern(pattern: str):
    # count number of consecutive of # into a list
    # ex: ##.#.##..# -> [2, 1, 2, 1]
    consecutive = []
    count = 0
    for char in pattern:
        if char == "#":
            count += 1
        else:
            if count > 0:
                consecutive.append(count)
                count = 0
    if count > 0:
        consecutive.append(count)
    return consecutive
```

```python
results = []
for idx, (p,v) in enumerate(zip(patterns, values)):
    print(idx,p,v)
    p = f"{p}?{p}?{p}?{p}?{p}"
    v = v*5
    solution_cnt = 0
    # solve for one?
    possible_locations = []
    wells_cnt = 0
    total_wells = sum(v)
    # 1. get possible locations
    for idx, char in enumerate(p):
        if char == '#':
            wells_cnt += 1
        if char == '?':
            possible_locations.append(idx)
    # 2. get combinations
    remaining_wells = total_wells - wells_cnt
    print("total", total_wells, "remain", remaining_wells, "possible", possible_locations)
    comb = itertools.combinations(possible_locations, remaining_wells)
    for c in comb:
        current_arr = [i for i in p]
        for idx in c:
            current_arr[idx] = '#'
        # check if matches
        current_v = build_value_from_pattern("".join(current_arr))
        if current_v == v:
            solution_cnt += 1
    results.append(solution_cnt)
results
```

```python
def build_value_from_pattern(pattern: str):
    # count number of consecutive of # into a list
    # ex: ##.#.##..# -> [2, 1, 2, 1]
    consecutive = []
    count = 0
    for char in pattern:
        if char == "#":
            count += 1
        else:
            if count > 0:
                consecutive.append(count)
                count = 0
    if count > 0:
        consecutive.append(count)
    return consecutive
```

```python
def find_possible_combinations(p, v):

    solution_cnt = 0
    # solve for one?
    possible_locations = []
    wells_cnt = 0
    total_wells = sum(v)
    # 1. get possible locations
    for idx, char in enumerate(p):
        if char == '#':
            wells_cnt += 1
        if char == '?':
            possible_locations.append(idx)
    # 2. get combinations
    remaining_wells = total_wells - wells_cnt
    # print("remaining_wells", remaining_wells)
    if remaining_wells < 0:
        return 0
    comb = itertools.combinations(possible_locations, remaining_wells)
    for c in comb:
        current_arr = [i for i in p]
        for idx in c:
            current_arr[idx] = '#'
        # check if matches
        current_v = build_value_from_pattern("".join(current_arr))
        if current_v == v:
            solution_cnt += 1
    return solution_cnt
```

```python
def build_cnt_dict(pattern, values):
    return {
        # 
        '0A#': find_possible_combinations(pattern + "#", values),
        "0A.": find_possible_combinations(pattern + ".", values),
        # 
        '.A.': find_possible_combinations("." + pattern + ".", values),
        '.A#': find_possible_combinations("." + pattern + "#", values),
        '#A.': find_possible_combinations("#" + pattern + ".", values),
        '#A#': find_possible_combinations("#" + pattern + "#", values),
        # 
        '.A0': find_possible_combinations("." + pattern, values),
        '#A0': find_possible_combinations("#" + pattern, values),
    }
```

```python
def build_cnt_type_from_str_rep(str_rep: str):
    # 0000 -> 0A. .A. .A. .A. .A0
    # 1000 -> 0A# .A. .A. .A. .A0
    # 0110 -> 0A. .A# #A# #A. .A0
    current = [['0','A', '.'], ['.','A', '.'], ['.','A', '.'], ['.','A', '.'], ['.','A', '0']]
    for idx, char in enumerate(str_rep):
        if char == '1':
            current[idx][2] = '#'
            current[idx+1][0] = '#'
    
    return ["".join(i) for i in current]
```

```python
queue = ['A', 'B']

def expand(s):
    if s[-1] == 'A' or s[-1] == 'a':
        return [s+'a', s+'b']
    elif s[-1] == 'B' or s[-1] == 'b':
        return [s+'A', s+'B']

finished_expand = []
while len(queue) > 0:
    current = queue.pop(0)
    expanded = expand(current)
    for i in expanded:
        if len(i) >= 5:
            finished_expand.append(i)
        else:
            queue.append(i)
```
