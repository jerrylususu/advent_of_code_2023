```python
with open("day19.txt", "r") as f:
    lines = [line.strip() for line in f.readlines()]
```

```python
lines
```

```python
workflows = {}
artifacts = []

for line in lines:
    if line == "":
        continue
    if not line.startswith("{"):
        # workflow
        workflow_name, workflow_step_str = line.replace("}","").split("{")
        workflow_steps = workflow_step_str.split(",")
        steps = []
        for step in workflow_steps:
            if ":" not in step:
                dim, comp, val, action = "always", "always", "always", step
            else:
                condition, action = step.split(":")
                dim, comp, val = condition[0], condition[1], int(condition[2:])
            steps.append((dim, comp, val, action))    
        workflows[workflow_name] = steps
    else:
        # artifact
        line = line.replace("{","").replace("}","")
        values = line.split(",")
        artifact = {}
        for value in values:
            dim, val = value.split("=")
            val = int(val)
            artifact[dim] = val
        artifacts.append(artifact)
```

```python
workflows
```

```python
def pass_artifact_through_workflow(artifact, workflows):
    current_workflow = "in"
    while current_workflow != "A" and current_workflow != "R":
        steps = workflows[current_workflow]
        for step in steps:
            dim, comp, val, action = step
            if dim == "always":
                current_workflow = action
                break
            else:
                dim_val = artifact[dim]
                if comp == '>':
                    if dim_val > val:
                        current_workflow = action
                        break
                elif comp == '<':
                    if dim_val < val:
                        current_workflow = action
                        break
                # no match, continue to next step
    return current_workflow
```

```python
accepted_artifacts = []
for artifact in artifacts:
    if pass_artifact_through_workflow(artifact, workflows) == "A":
        accepted_artifacts.append(artifact)
```
